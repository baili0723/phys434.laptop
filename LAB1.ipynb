```python
#import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading data
data = pd.read_csv('HTRU_2.csv',names=['fea1','fea2', 'fea3','fea4','fea5','fea6','fea7','fea8','fea9'],index_col=None)
print(data.head())

#seperate pulsar and noise
class_labels = data.iloc[:,-1]
pulsar = data[class_labels == 1]
noise = data[class_labels == 0]
```

```python
#create histogram
#featrue 1
#class1
plt.hist(pulsar['fea1'])
plt.show()
#calss 0
plt.hist(noise['fea1'])
plt.show()
#together
plt.hist(pulsar['fea1'])
plt.hist(noise['fea1'])
plt.show()
#same bin and range
plt.hist(pulsar['fea1'],bins=100,range=(0,200))
plt.hist(noise['fea1'],bins=100,range=(0,200))
plt.show()
#normailized
plt.hist(pulsar['fea1'],bins=100,range=(0,200),density=True)
plt.hist(noise['fea1'],bins=100,range=(0,200),density=True)
plt.show()
#different color line style
plt.hist(pulsar['fea1'],bins=100,range=(0,200),density=True,color='blue',label='pulsar',linestyle='--')
plt.hist(noise['fea1'],bins=100,range=(0,200),density=True,color='red',label='noise', linestyle=':')
#add label
plt.xlabel('histrgram of Mean of the integrated profile')
plt.ylabel('Frequency')
mean_p_1=pulsar['fea1'].mean()
std_p_1=pulsar['fea1'].std()
mean_n_1=noise['fea1'].mean()
std_n_1=noise['fea1'].std()

plt.legend()
plt.show()
```

```python
total_observations = len(data)
pulsar_candidates = len(pulsar)
probability_a = pulsar_candidates / total_observations
print(probability_a)

# Calculate P(B) - Probability that an observation has Mean of Integrated Profile < 80
mean_profile_lt_80 = len(data[data['fea1'] < 80])
probability_b = mean_profile_lt_80 / total_observations
print(probability_b)

# Calculate P(A ⋂ B) - Probability that an observation satisfies both conditions
both_conditions = len(data[(data['fea9'] == 1) & (data['fea1'] < 80)])
probability_a_and_b = both_conditions / total_observations
print(probability_a_and_b)
```

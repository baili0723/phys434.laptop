## **1.Variable of Transformation**

```python
x = np.random.uniform(0,1,1000)

u = x ** 2

#print(x[789]**2 , u[789])

# calculate g(u)
# u = x**2
# x(u) = sqrt(u)
# x'(u) = 1/sqrt(u) * 1/2
# g(u) = 1/ ( 2 * sqrt(u))

def g(u):
    return 1 / (2 * np.sqrt(u))

plt.hist(u, bins=100, density=True, alpha=0.6, color='g', label='u')
u_values = np.linspace(0.01, 1, 1000)
plt.plot(u_values, g(u_values), 'r', lw=2, label='Analytical PDF: g(u)')
```

# **3.Central Limit Theorem**

```python
from scipy.stats import poisson

# Lists to store means and standard deviations for each sample
mean = []
std = []

#def function
def event(N):
  mean = []
  std = []
  for i in range(N):
    # Generate a sample of 5 events from a Poisson distribution
    sample = poisson.rvs(mu=2, size=5)

    # Calculate the mean and standard deviation of the sample
    a = np.mean(sample)
    b = np.std(sample)

    # Append the results to the lists
    mean.append(a)
    std.append(b)
  return mean,std

# N=10
results = event(10)
mu5 = results[0]
lambda5 = results[1]

plt.hist(mu5, bins=50, color='g', label='u')
plt.show
```
